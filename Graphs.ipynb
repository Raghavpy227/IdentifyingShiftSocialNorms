```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import os

path=r"C:\Users\py22715\OneDrive - University of Bristol\stats"
df_nocontext=pd.read_csv(os.path.join(path,"NoContextStats.csv"))
df_context=pd.read_csv(os.path.join(path,"Context_Stats.csv"))
```

```python
plt.title("Training Losses by Epochs [Without Context]")
plt.xlabel("Epoch")
plt.ylabel("Training Error")
sns.lineplot(data=df_nocontext["Training_Loss_BERT"],label='BERT')
sns.lineplot(data=df_nocontext["Training_Loss_PromptModel"],label='PromptModel')
```

```python
plt.title("F1 Micro Scores by Epochs [Without Context]")
plt.xlabel("Epoch")
plt.ylabel("F1_Scores (Micro)")
sns.lineplot(data=df_nocontext["F1_Validation_BERT"],label='BERT')
sns.lineplot(data=df_nocontext["F1_Validation_PromptModel"],label='PromptModel')
```

```python
plt.title("Training Time by Epochs [Without Context]")
plt.xlabel("Epoch")
plt.ylabel("Training Time")
sns.lineplot(data=df_nocontext["Training_Time_BERT"],label='BERT')
sns.lineplot(data=df_nocontext["Training_Time_PrompyModel"],label='PromptModel')
```

```python
plt.title("Training Losses by Epochs [Context]")
plt.xlabel("Epoch")
plt.ylabel("Training Error")
sns.lineplot(data=df_context["Training_Loss_BERT"],label='BERT')
sns.lineplot(data=df_context["Training_Loss_PromptModel"],label='PromptModel')
```

```python
plt.title("F1 Micro Scores by Epochs [Context]")
plt.xlabel("Epoch")
plt.ylabel("F1_Scores (Micro)")
sns.lineplot(data=df_context["F1_Validation_BERT"],label='BERT')
sns.lineplot(data=df_context["F1_Validation_PromptModel"],label='PromptModel')
```

```python
plt.title("Training Time by Epochs [Context]")
plt.xlabel("Epoch")
plt.ylabel("Training Time")
sns.lineplot(data=df_context["Training_Time_BERT"],label='BERT')
sns.lineplot(data=df_context["Training_Time_PrompyModel"],label='PromptModel')
```
